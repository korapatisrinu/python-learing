```python
def check_num(n):
    if n>0:
        print("Positive")
    else:
        print("Negative")
check_num(-4)
```

```python
def table(n):
    for i in range(1,5):
        print(n*i)
table(2)
```

```python
def count_vowles(s):
    count=0
    for i in s:
        if i in "aeiouAEIOU":
            count+=1
            print(count)
count_vowles("python")
```

```python
def sum_n(n):
    total=0
    for i in range(1,n+1):
        total+=i
    print(total)

sum_n(5)
```

```python
def fac(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    print(fact)
```

```python
fac(5)
```

```python
def prime(n):
    if n < 2:
        return "Not Prime"
    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"
    return "Prime"

print(prime(10))
```
